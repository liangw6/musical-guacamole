```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
input_file = "bell_sound.wav"
```

```python
fs, data =  wavfile.read(input_file)
print(data)
print(data.shape)
```

```python
channel_0 = data[:,0]
channel_1 = data[:,1]
plt.plot(channel_0, label='channel 0', alpha=0.5)
plt.plot(channel_1, label='channel 1', alpha=0.5)
plt.legend()
```

```python
# create sin wave
sample_rate = 44100
freq = 400
samples = 44100
amplitude = 100


x = np.arange(samples)
y = amplitude * np.sin(2 * np.pi * freq * x / sample_rate)
plt.plot(y)
plt.xlim(0, 1024)
```
