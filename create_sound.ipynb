```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
input_file = "Scarborough-Fair.wav"
```

```python
fs, data =  wavfile.read(input_file)
print(data)
print(data.shape)
```

```python
channel_0 = data[:,0]
channel_1 = data[:,1]
plt.plot(channel_0[:1024], label='channel 0')
plt.plot(channel_1[:1024], label='channel 1')
plt.legend()
plt.savefig('before.png')
```

```python
# create sin wave
sample_rate = 44100
freq = 400
samples = 44100
amplitude = 100


x = np.arange(samples)
y = amplitude * np.sin(2 * np.pi * freq * x / sample_rate)
plt.plot(y)
plt.xlim(0, 1024)
plt.savefig('addition_png')
```

```python
combined_0 = channel_0[:44100] + y
combined_1 = channel_1[:44100] + y
plt.plot(combined_0[:1024], label='channel 0')
plt.plot(combined_1[:1024], label='channel 1')
plt.savefig('after.png')
```
